# Definir agente RAG

```python
max_iterations = 3
```

```python
import time
```

```python
# Paso 1: Definir las funciones de cada etapa del pipeline

def retrieve_context(question: str) -> dict:
    """
    Consulta el vector store para obtener los fragmentos más relevantes a la pregunta.
    Imprime información de debug sobre los documentos recuperados.
    """
    similar_docs = vector_store.similarity_search(question, k=1)

    # Debug: imprimir información de los documentos recuperados
    #print("DEBUG: Documentos recuperados:")
    #for i, doc in enumerate(similar_docs):
    #    snippet = doc.page_content[:100].replace("\n", " ")
    #    print(f"  Documento {i+1}: {snippet}...")

    time.sleep(120)

    # Concatenar el contenido para formar el contexto
    context = "\n".join([doc.page_content for doc in similar_docs])
    #print("DEBUG: Contexto concatenado (primeros 200 caracteres):", context[:200], "...")

    return {"question": question, "context": context}

def generate_answer(inputs: dict) -> dict:
    """
    Combina la pregunta y el contexto en un prompt y utiliza el LLM para generar una respuesta.
    Imprime el prompt y la respuesta para depuración.
    """
    prompt = (
        f"Contexto:\n{inputs['context']}\n\n"
        f"Pregunta:\n{inputs['question']}\n\n"
        "Utiliza la información del contexto para responder de manera clara, detallada y precisa. "
        "Asegúrate de identificar y mencionar los puntos clave del contexto que fundamentan tu respuesta, "
        "proporcionando datos o referencias cuando sea posible."
    )
    #print("DEBUG: Prompt para generación de respuesta:\n", prompt)

    # Se asume que 'llm' está configurado y tiene el método 'generate'
    answer_obj = llm.generate_content(prompt)
    answer = answer_obj.text
    #print("DEBUG: Respuesta generada:", answer)

    return {"question": inputs["question"], "context": inputs["context"], "answer": answer}

def validate_answer(inputs: dict) -> dict:
    """
    Evalúa la respuesta generada, indicando en qué partes del contexto se ha basado y por qué la respuesta es válida.
    Imprime el prompt de validación y la reflexión generada.
    """
    reflection_prompt = (
        "Analiza detenidamente la siguiente información. Luego, evalúa la respuesta generada considerando lo siguiente:\n"
        "1. ¿La respuesta aborda todos los aspectos relevantes de la pregunta?\n"
        "2. ¿Qué partes específicas del contexto respaldan la respuesta?\n"
        "3. ¿Por qué la respuesta es válida o en qué aspectos podría mejorarse?\n"
        "Si identificas áreas que necesitan mayor detalle o precisión, indica qué información adicional sería necesaria.\n\n"
        f"Contexto:\n{inputs['context']}\n\n"
        f"Pregunta:\n{inputs['question']}\n\n"
        f"Respuesta:\n{inputs['answer']}\n\n"
        "Proporciona una reflexión detallada y crítica:"
    )
    #print("DEBUG: Prompt para validación de respuesta:\n", reflection_prompt)

    reflection_obj = llm.generate_content(reflection_prompt)
    reflection = reflection_obj.text
    #print("DEBUG: Reflexión generada:", reflection)

    return {
        "question": inputs["question"],
        "context": inputs["context"],
        "answer": inputs["answer"],
        "reflection": reflection
    }

def reflection_indicates_improvement_needed(reflection: str) -> bool:
    """
    Determina si la reflexión sugiere que la respuesta es deficiente o incompleta.
    Se buscan palabras clave que indiquen la necesidad de mejora.
    """
    keywords = [
        "deficiencia", "incompleta", "mejorar", "insuficiente", "falta",
        "no es lo suficientemente buena", "podría mejorar", "necesita mayor detalle", "no aborda"
    ]
    for keyword in keywords:
        if keyword.lower() in reflection.lower():
            return True
    return False

def improve_answer(inputs: dict, reflection: str) -> dict:
    """
    Utiliza la retroalimentación proporcionada para generar una respuesta mejorada que aborde los puntos débiles señalados.
    """
    prompt = (
        f"Se te ha proporcionado la siguiente retroalimentación sobre la respuesta anterior:\n"
        f"{reflection}\n\n"
        "Utilizando esta retroalimentación, genera una respuesta mejorada que aborde los puntos débiles mencionados, "
        "proporcionando mayor detalle, precisión y claridad.\n\n"
        f"Contexto:\n{inputs['context']}\n\n"
        f"Pregunta:\n{inputs['question']}\n\n"
        "Respuesta mejorada:"
    )
    #print("DEBUG: Prompt para generar respuesta mejorada:\n", prompt)
    improved_answer_obj = llm.generate_content(prompt)
    improved_answer = improved_answer_obj.text
    #print("DEBUG: Respuesta mejorada generada:", improved_answer)
    return {"question": inputs["question"], "context": inputs["context"], "answer": improved_answer}


def rag_pipeline(question: str) -> dict:
    # 1. Recuperar contexto
    context_data = retrieve_context(question)

    # 2. Generar respuesta
    answer_data = generate_answer(context_data)

    # 3. Validar la respuesta
    reflection_data = validate_answer(answer_data)
    iteration = 0
    while iteration < max_iterations and reflection_indicates_improvement_needed(reflection_data["reflection"]):
        #print(f"DEBUG: Iteración {iteration+1}: La reflexión indica que la respuesta necesita mejoras.")
        #print("Generando una respuesta mejorada...")
        improved_answer_data = improve_answer(answer_data, reflection_data["reflection"])
        # Actualizar la respuesta con la respuesta mejorada y volver a validar
        answer_data = improved_answer_data
        reflection_data = validate_answer(answer_data)
        iteration += 1

    return reflection_data
```

```python
# Paso 3: Ejecutar el pipeline con una pregunta de ejemplo

input_question = "¿Cuantos tipos de cesion hay?¿Cuales son?"
result = rag_pipeline(input_question)

print("\n----- RESULTADO FINAL -----")
print("Pregunta:", result["question"])
print("Respuesta generada:", result["answer"])
print("Reflexión sobre la respuesta:", result["reflection"])
```
